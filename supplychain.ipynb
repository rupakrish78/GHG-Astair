```python
import pandas as pd

data = pd.read_csv ("SupplyChainGHGEmissionFactors_v1.2_NAICS_CO2e_USD2021.csv")
data.head()
```

```python
import altair as alt

bargraph_unfiltered = alt.Chart (data).mark_bar().encode (
    alt.Y ("2017 NAICS Title", sort ='-x'),
    x = "Supply Chain Emission Factors with Margins"
)

text = bargraph_unfiltered.mark_text (baseline = "middle", align = "left", color = "black").encode (text = "Supply Chain Emission Factors with Margins")

final_unfiltered = bargraph_unfiltered + text
final_unfiltered.save("bargraph_unfiltered.html", embed_options = {'renderer' : 'svg'})
```

```python
import altair as alt

data_filtered = data[data["Supply Chain Emission Factors with Margins"] < 5]

bargraph_filtered = alt.Chart (data_filtered).mark_bar().encode (
    alt.Y ("2017 NAICS Title", sort ='-x'),
    x = "Supply Chain Emission Factors with Margins"
)

text = bargraph_filtered.mark_text (baseline = "middle", align = "left", color = "black").encode (text = "Supply Chain Emission Factors with Margins")

final_filtered = bargraph_filtered + text
final_filtered.save("bargraph_filtered.html", embed_options = {'renderer' : 'svg'})
```
